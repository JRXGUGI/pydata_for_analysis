### Figure 和 Subplot
创建图表对象主要有三种方法
- 1、单个创建 使用fig.add_subplot对象,首先要生成fig对象
        ax1 = fig.add_subplot(2,2,1)

- 2、单个创建，使用 plt.subplot(321)，之间plt调用，调用者不同
        ax1 = plt.subplot(321)  

- 3、创建子图对象数组（推荐）：
        fig, axes = plt.subplots(2,2) # axes是一个2*2的二维数组

````python
plt.style.use('ggplot') 
# 1 matplotlib的图像都位于Figure对象中
fig = plt.figure(figsize=(10,8))

# 2 使用add_subplot创建一个或者多个subplot
ax1 = fig.add_subplot(2,2,1)  # add_subplot 是创建子图对象
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
# 方法二：或者直接使用plt.subplots函数，直接创建figure对象和axes对象数组
# fig, axes = plt.subplots(2,2)
# 方法三：直接创建一个axes对象 该方法与add_subplot函数类似
# ax1 = plt.subplot(321)

# 3 绘制图形 如果不指定，直接在最后一个axes对象上绘制
plt.plot(np.random.randn(50).cumsum(), 'k--');

# 4 指定axes对象绘制
ax1.hist(np.random.randn(100), bins=20, color='k',alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
````

````python
# 使用subplots方法更便捷的创建图表对象
fig, axes = plt.subplots(nrows=2, ncols=3,figsize=(10,8) )
axes
axes[0,1].hist(np.random.randn(100), bins=20, color='k',alpha=0.3)
axes[1,2].scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
axes[1,0].plot(np.random.randn(50).cumsum(), 'k--');
````

### 调整subplot周围的间距
主要使用subplots_adjust方法；
```python
subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)
    参数：
    ------
    wspace/hspace:用于控制宽度和高度的百分比
    left/bottom/right/top 边界距离百分比，以左边轴和底边轴为标准
```

````python
# 1 子图间距缩小为0 
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(10,8))
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=30, color='red',alpha=0.5)
plt.subplots_adjust(wspace=0, hspace=0);
````

### 设置标题、轴标签、刻度以及刻度标签
> 子图对象ax有相关的设置方法，组合图对象使用plt下的相关方法

````python
# 注意 该部分只是设置一个子图的方法，如果是组合图，需参考plt下的相关方法
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())

# 设置轴标签
ticks = ax.set_xticks([0, 250, 500, 750, 1000]) # 标签位置
labels = ax.set_xticklabels(['one', 'two', 'three', 'four','five'],rotation=30, fontsize='small') #标签内容 角度 字体大小
ax.set_xlabel('Stages')  # 设置轴标签名称、

# 设置标题
# ax.set_title('My first matplotlib plot')

## 批量设置
props = {
    'title': 'My first matplotlib plot',
    'ylabel': 'YStages'

}
ax.set(**props); # 使用set函数可以进行批量设置 
````

### 注解以及在Subplot上绘图
注解和文字可以通过text、arrow和annotate函数进行添加。text可以将文本绘制在图表的指定坐标(x,y)

````python
from datetime import datetime
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
data = pd.read_csv('examples/spx.csv', index_col=0, parse_dates=True)

spx = data['SPX']
spx.plot(ax=ax, style='k-')

crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]

# 添加注解
for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 75),
                 xytext=(date, spx.asof(date) + 225),
                 arrowprops=dict(facecolor='red', headwidth=4,width=2,headlength=4),
                 horizontalalignment='left', verticalalignment='top')


# Zoom in on 2007-2010
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])
ax.set_title('Important dates in the 2008-2009 financial crisis');
````

````python
# 绘制图形
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='k', alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color='b', alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
            color='g', alpha=0.5)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon);
````

### 线型图
Series和DataFrame都有一个用于生成各类图表的plot方法。默认情况下，它们所生成的是线型图

````python
# help(pd.Series.plot)
s = pd.Series(np.random.randn(10).cumsum(),index=np.arange(0, 100, 10))
fig,axes = plt.subplots(2,1,figsize=(6,4))
s.plot(rot=30, grid=False, use_index=False, ax=axes[0])
s.map(lambda x:round(x,0)).value_counts().plot.bar(ax=axes[1]);
````
